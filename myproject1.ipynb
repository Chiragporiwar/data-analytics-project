```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))
fig.suptitle("Sales with different shipping modes and segments", fontsize=16)

# Plot for Ship Mode vs Sales
sns.barplot(x='Ship Mode', y='Sales', data=df, ax=axes[0, 0], dodge=False)

# Plot for Ship Mode vs Sales (lineplot)
sns.lineplot(x='Ship Mode', y='Sales', data=df, ax=axes[0, 1])

# Plot for Segment vs Sales
sns.barplot(x='Segment', y='Sales', data=df, ax=axes[1, 0])

# Plot for Segment vs Sales (lineplot)
sns.lineplot(x='Segment', y='Sales', data=df, ax=axes[1, 1])

plt.tight_layout()
plt.show()
```

```python
# Segment customers based on total sales
high_value_customers = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).index[:3]

# Filter DataFrame for high-value customers
high_value_df = df[df['Segment'].isin(high_value_customers)]

segment_analysis = high_value_df.groupby('Segment').agg({
    'Sales': 'mean',          
    'Quantity': 'mean',       
    'Discount': 'mean',  
    'Profit': 'mean' 
})

# Plotting the characteristics of each segment
segment_analysis.plot(kind='bar', figsize=(8, 6))
plt.title('Segment Analysis')
plt.xlabel('Segment')
plt.ylabel('Average Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metric')
plt.show()

# Segment customers based on total sales
high_value_customers = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).index[:3]

# Filter DataFrame for high-value customers
high_value_df = df[df['Segment'].isin(high_value_customers)]

segment_analysis = high_value_df.groupby('Segment').agg({
    'Sales': 'mean',          # Average sales
    'Quantity': 'mean',       # Average quantity
    'Discount': 'mean',       # Average discount
    'Profit': 'mean'          # Average profit
})

print(segment_analysis)
```

```python
# Calculate total sales for each sub-category within each category
subcat_sales = df.groupby(['Category', 'Sub-Category'])['Sales'].sum()

# Sort the sub-categories based on total sales and select the top 10 for each category
top_10_subcat = subcat_sales.groupby('Category').nlargest(10)

# Plot the top 10 sales entries for each category
plt.figure(figsize=(10, 6))
for category, data in top_10_subcat.groupby('Category'):
    plt.bar(data.index.get_level_values('Sub-Category'), data.values, label=category)

plt.title('Top 10 Sales by Sub-Category in Each Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
